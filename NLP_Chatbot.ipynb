```python
import random
import re
import string
import nltk
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resource packages
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt_tab')

################################################################################
# Initialize the Lemmatizer to find the base/root form of words
lemmatizer = WordNetLemmatizer()

################################################################################
# Define chatbot knowledge base (Intents, Patterns, and Responses)
intents = [
    {
        "tag": "greeting",
        "patterns": ["hi", "hello", "hey", "good morning", "greetings"],
        "responses": ["Hello! How can I help you today?", "Hi there!", "Hey! What's up?"]
    },
    {
        "tag": "goodbye",
        "patterns": ["bye", "see you later", "goodbye", "exit", "quit"],
        "responses": ["Goodbye! Have a great day.", "See you later!", "Bye! Take care."]
    },
    {
        "tag": "thanks",
        "patterns": ["thanks", "thank you", "that's helpful", "awesome"],
        "responses": ["You're very welcome!", "Anytime!", "Glad I could help!"]
    },
    {
        "tag": "identity",
        "patterns": ["what is your name", "who are you", "what are you"],
        "responses": ["I am a simple NLP chatbot.", "You can call me Python Bot!"]
    }
]
################################################################################
def clean_and_tokenize(text):
    """
    Cleans user input by removing punctuation, converting to lowercase,
    tokenizing into words, and lemmatizing them.
    """
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize words and lowercase them
    words = nltk.word_tokenize(text.lower())
    # Lemmatize each word (e.g., "greetings" becomes "greeting")
    return [lemmatizer.lemmatize(word) for word in words]

################################################################################
def get_response(user_input):
    """
    Matches lemmatized user words against the intent patterns.
    """
    user_words = clean_and_tokenize(user_input)

    # Loop through each intent to see if any pattern word matches user input
    for intent in intents:
        for pattern in intent["patterns"]:
            pattern_words = clean_and_tokenize(pattern)
            # Check if all words of a pattern are present in user text
            if all(word in user_words for word in pattern_words):
                return random.choice(intent["responses"])

    return "I'm sorry, I didn't quite understand that. Could you try rephrasing?"

################################################################################
# Main Chat Loop
def start_chat():
    print("🤖 Bot: Hello! I am an NLP Chatbot. (Type 'exit' or 'bye' to quit)\n")

    while True:
        user_input = input("You: ")

        # Check if user wants to end the session
        if user_input.lower() in ['exit', 'bye', 'quit']:
            print("🤖 Bot: Goodbye! Have a wonderful day.")
            break

        response = get_response(user_input)
        print(f"🤖 Bot: {response}\n")

################################################################################
if __name__ == "__main__":
    start_chat()
```
