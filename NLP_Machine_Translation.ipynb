```python
# You will need to install the googletrans-py library:
! pip install -q googletrans-py

from googletrans import Translator

# 1. Initialize the translator object
translator = Translator()
```

```python
# 2. Define a phrase to translate into multiple target languages
source_phrase = "Artificial intelligence bridges the gap between different human cultures."

# Define target language codes: 'fr' (French), 'de' (German), 'ja' (Japanese)
languages = {
    'fr': 'French',
    'de': 'German',
    'ja': 'Japanese'
}

print("--- MULTI-LANGUAGE API TRANSLATION ---")
print(f"Source Text: {source_phrase}\n")
```

```python
# 3. Loop through and translate dynamically
for lang_code, lang_name in languages.items():
    # The API will automatically detect the source language (English)
    translation = translator.translate(source_phrase, dest=lang_code)

    print(f"{lang_name} ({lang_code}): {translation.text}")
```
